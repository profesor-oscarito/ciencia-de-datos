# Data Science Pipeline (flujo completo)
## Raw Data (Datos crudos)
> 👉 Datos originales, sin procesar
- CSV, Excel, APIs, bases de datos
- Pueden tener:
    - errores
    - valores nulos 
    - datos inconsistentes

> **Ejemplo**

nombre | edad | sueldo
Juan   | 25   | 1000000
Ana    | NaN  | ?


# Data Cleaning (Limpieza de datos)
> 👉 Se corrigen problemas
- Eliminar duplicados
- Manejar valores nulos
- Corregir formatos
- Convertir tipos de datos
> Utilizas:
- Panda y numpy

# Clean Data (Datos limpios) 
> 👉 Datos listos para trabajar
- Sin errores
- Consistentes
- Estructurados
> ✔ Este es el resultado de la limpieza


Raw Data 
   ↓
