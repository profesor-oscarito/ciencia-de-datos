```python
# EJEMPLO EN PYTHON

import pandas as pd

# Crear un DataFrame ejemplo con problemas
df = pd.DataFrame({
    'Nombre': ['Ana', 'Luis', 'Ana', 'Carlos', None],
    'Edad': [23, None, 23, 45, 31],
    'Nota': [5.5, 6.0, 5.5, 7.0, None]
})

print(df)
print("*" * 40)
# Eliminar duplicados
df_clean = df.drop_duplicates()

# Imputar valores faltantes con la media
df_clean['Edad'] = df_clean['Edad'].fillna(df_clean['Edad'].mean()) 
# Imputar valores faltantes con la media
df_clean['Nota'] = df_clean['Nota'].fillna(df_clean['Nota'].mean())

print(df_clean)
```

```python
df.info() # Información general del DataFrame
```

```python
df.isna().sum() # Contar valores faltantes por columna
```

> Esto es lo que significa cada parte:

- count : Número de entradas no nulas (no faltantes). En este caso, 5 para ambas columnas.
- media : El valor promedio. Para Age, es 35. Para Salary, es 77.000.
- Desviación estándar (std) : Indica la dispersión de los datos. Un valor alto significa mayor variabilidad.
- min : El valor más pequeño de la columna.
- 25%, 50%, 75% (percentiles/cuartiles) :
    - 25% : El primer cuartil: el 25% de los datos se encuentra por debajo de este valor.
    - 50% : La mediana (valor central).
    - 75% : El tercer cuartil: el 75% de los datos se encuentra por debajo de este valor.
- máx : El valor más alto de la columna.
