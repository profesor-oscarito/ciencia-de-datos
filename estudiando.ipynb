```python
# Importa la clase IterativeImputer, que es el algoritmo que rellena valores faltantes (NaN)
# Este método usa el algoritmo MICE (Multivariate Imputation by Chained Equations)

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

import pandas as pd
import numpy as np  # Agregar esto si no está arriba

X = pd.DataFrame({
    'edad': [25, 30, np.nan, 45],
    'salario': [30000, np.nan, 45000, 60000],
    'experiencia': [2, 5, 7, np.nan]
})

print(X)
imputer = IterativeImputer(max_iter=10)
X_imputed = imputer.fit_transform(X)

print(X_imputed) # Imprime el DataFrame con los valores faltantes rellenados
```

```python
import numpy as np

# 1. Crear un arreglo bidimensional (Matriz 2x3)
# 2 filas, 3 columnas
matriz = np.array([[1, 2, 3], 
                   [4, 5, 6]])

print("--- Matriz Original (2x3) ---")
print(matriz)

# 2. Acceder a elementos [fila, columna]
print("\nElemento en fila 1, columna 2:", matriz[1, 2]) # Resultado: 6

# 2. Acceder a elementos [fila, columna]
print("\nElemento en fila 1, columna 1:", matriz[1, 1]) # Resultado: 3

# 3. Operaciones aritméticas (vectorizadas)
print("\n--- Matriz multiplicada por 2 ---")
print(matriz * 2)

# 4. Cambiar la forma del arreglo (reshape)
# De 2x3 a 3x2
matriz_redimensionada = matriz.reshape(3, 2)
print("\n--- Matriz redimensionada (3x2) ---")
print(matriz_redimensionada)

# 5. Obtener dimensiones
print("\nForma del arreglo (shape):", matriz.shape) # Resultado: (2, 3)

```
