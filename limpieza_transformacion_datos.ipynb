# Limpieza y Transformación de Datos con Python

Ejemplo práctico para Ciencia de Datos

```python
import pandas as pd
import numpy as np
```

## 1. Crear datos con errores

```python
data = {
    "nombre": ["Juan", "María", "Pedro", "Ana", "Luis", None],
    "edad": [25, 30, None, 22, 40, 35],
    "profesion": ["Ingeniero", "doctor", "Profesor", "Ingeniero", None, "doctor"],
    "sueldo": ["1000000", "1200000", None, "950000", "1100000", "1300000"]
}

df = pd.DataFrame(data)
print(df)
```

## 2. Exploración de datos

```python
df.info() # explora la estructura del DataFrame, mostrando el número de entradas, el número de columnas, los nombres de las columnas, el tipo de datos de cada columna y la cantidad de valores no nulos en cada columna. Esta información es útil para entender la composición del DataFrame y detectar posibles problemas con los datos, como valores faltantes o tipos de datos incorrectos.
print("*" * 50)
df.isnull().sum() #

df
```

## 3. Limpieza de datos

```python
# Eliminar nulos en nombre
df = df.dropna(subset=["nombre"])

# Rellenar edad con promedio
df["edad"] = df["edad"].fillna(df["edad"].mean())

# Rellenar profesión
df["profesion"] = df["profesion"].fillna("Desconocido")

```

## 4. Transformación de datos

```python
# Convertir sueldo a numérico
df["sueldo"] = pd.to_numeric(df["sueldo"], errors="coerce")

# Rellenar nulos con mediana
df["sueldo"] = df["sueldo"].fillna(df["sueldo"].median())

# Estandarizar texto
df["profesion"] = df["profesion"].str.lower()

# Crear categoría de edad
df["categoria_edad"] = pd.cut(df["edad"],
                             bins=[0, 30, 50, 100],
                             labels=["Joven", "Adulto", "Mayor"])

df
```

## 5. Análisis simple

```python
# explicar por linea
df.groupby("profesion")["sueldo"].mean()
```
